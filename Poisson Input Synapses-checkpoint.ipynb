```python
from brian2 import *
import numpy as np
%matplotlib inline
```

```python
v_th = 20*mV # Spiking Schwelle

c_acd = 10*nfarad/mm**2 # specific membrane capacitance aus dem Buch/Paper
r_acd = 1*Mohm*mm**2 # specific membrane restistance aus dem Buch
c_soma = 10*nfarad/mm**2 # aus dem Paper Supplementary Table
r_soma = 1*Mohm*mm**2 # specific membrane restistance aus dem Buch
r_l = 2*kohm*mm # vielleicht wie im neuen Paper

d_soma = 30*um #aus Paper in Computational Modeling
A_soma = 3*pi*d_soma**2 # dreimal erhöhter Oberfläche um sich der Zellgröße von 12,500µm^2 anzunähern
A_acd = pi*1.4*um*150*um # Aus dem Paper in Modelling Annahme als Zylinder
 
C_acd = c_acd*A_acd
C_soma = c_soma*A_soma

tau_acd = r_acd*c_acd
tau_soma = r_soma*c_soma
tau_syn_ex = 2*ms
tau_syn_inh = 5*ms

reset='''
v_acd = 0*mV
v_soma = 0*mV
'''

R_l = 6.5*um*r_l / (pi * (0.7*um)**2) # 6.5 ist durchschnittlicher Abstand zwischen ACD und Soma
```

```python
def potential_dif_finder(M,N):
    potentials_soma = [M[i] for i in range(len(M)-1) if M[i+1] == 0]
    potentials_s =[value for value in potentials_soma if value != 0]
    potentials_acd = [N[i] for i in range(len(N)-1) if N[i+1] == 0]
    potentials_a =[value for value in potentials_acd if value != 0]
    potential_dif_before = [abs(potentials_s[i] - potentials_a[i]) for i in range(len(potentials_s))]
    potential_dif = mean(potential_dif_before)
    return potential_dif
```

# Membrane Potentials with Input at ACD

```python
eqs = '''
dv_acd/dt = -(1/tau_acd)*v_acd + 1/C_acd *I_acd + 1/(C_acd * R_l) *(v_soma-v_acd) : volt
dv_soma/dt = -1/(tau_soma)*v_soma + 1/C_soma *I_soma + 1/(C_soma * R_l)*(v_acd-v_soma) : volt
dI_acd/dt = -(1/tau_syn_ex)*I_acd : ampere
dI_soma/dt = -(1/tau_syn_ex)*I_soma : ampere 
'''
```

```python
start_scope()
v_th = 20*mV # Spiking Schwelle
defaultclock.dt = 0.01*ms
N = 20  # Amount of Neurons
num_inputs_ex = 1 
num_inputs_inh = 1
input_rate_ex = 1000*Hz
input_rate_inh = 1000*Hz
duration = 100*ms

AcD1 = NeuronGroup(N, eqs, threshold="v_acd > v_th", reset=reset, method='rk4')
AcD1.v_acd = 0*mV  #The following arbitrarly choosen inputs
AcD1.v_soma = 0*mV
AcD1.I_acd = 0.0*namp
AcD1.I_soma = 0.0*namp

PSyn_soma_ex = PoissonGroup(num_inputs_ex, rates=input_rate_ex)
PSyn_acd_ex = PoissonGroup(num_inputs_ex, rates=input_rate_ex)
PSyn_soma_inh = PoissonGroup(num_inputs_inh, rates=input_rate_inh)
PSyn_acd_inh = PoissonGroup(num_inputs_inh, rates=input_rate_inh)

S_acd_ex = Synapses(PSyn_acd_ex, AcD1, model='syn_weight : ampere', on_pre='I_acd_post += syn_weight')
S_acd_ex.connect()
S_acd_ex.syn_weight = 0.12*namp

# S_acd_inh = Synapses(PSyn_acd_inh, AcD1, model='syn_weight : ampere', on_pre='I_acd_post += syn_weight')
# S_acd_inh.connect()
# S_acd_inh.syn_weight = -0.0066*namp

# S_soma_ex = Synapses(PSyn_soma_ex, AcD1, model='syn_weight : ampere', on_pre='I_soma_post += syn_weight')
# S_soma_ex.connect()
# S_soma_ex.syn_weight = 0.03*namp

S_soma_inh = Synapses(PSyn_soma_inh, AcD1, model='syn_weight : ampere', on_pre='I_soma_post += syn_weight')
S_soma_inh.connect()
S_soma_inh.syn_weight = -0.06*namp


M = StateMonitor(AcD1, ('v_acd', 'v_soma'), record=True)
Spike = SpikeMonitor(AcD1)

run(duration)
```

```python
plot(M.t/ms, M.v_acd[0].T/mV, label="membrane potential acd")
plot(M.t/ms, M.v_soma[0].T/mV, label="membrane potential soma")
plt.title(label="Membrane potentials of postsynaptic Neuron")
plt.legend(loc="upper right")
xlabel('Time (ms)')
ylabel('different v_potentials (mV)');
```
